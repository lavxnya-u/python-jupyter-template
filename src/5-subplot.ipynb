### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_sizes = [40, 25, 20, 15]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y, color='blue', linewidth=2)
axs[0, 0].set_title('Line Chart', fontsize=12)

axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'orange'])
axs[0, 1].set_title('Bar Chart', fontsize=12)

axs[1, 0].scatter(scatter_x, scatter_y, color='purple', edgecolor='black')
axs[1, 0].set_title('Scatter Plot', fontsize=12)

axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart', fontsize=12)

plt.tight_layout()
plt.show()
```
